```python
import pandas as pd
import geopandas as gpd
from sqlalchemy import create_engine, text
import shapely.wkb
import folium

# Connection
DB_URL = "postgresql://user:password@localhost:5432/osm_sg"
engine = create_engine(DB_URL)

def map_anything():
    # 1. Get the column name (confirmed as 'geom' in your last run)
    geom_col = 'geom' 

    # 2. Query data - keeping it to 1000 for a fast demo
    sql = f"""
    SELECT name, fclass, ST_AsBinary(ST_Transform({geom_col}, 4326)) AS geom_wkb 
    FROM raw_osm_data 
    WHERE name IS NOT NULL 
    LIMIT 1000
    """
    
    print("Fetching and centering Singapore POIs...")
    df = pd.read_sql(sql, engine)
    
    if df.empty:
        print("No data found. Check if your import finished.")
        return

    # 3. Convert to GeoDataFrame
    df['geometry'] = df['geom_wkb'].apply(lambda x: shapely.wkb.loads(bytes(x)))
    gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

    # 4. Map setup centered on Singapore [cite: 113]
    m = folium.Map(location=[1.3521, 103.8198], zoom_start=12, tiles="CartoDB positron")

    # 5. THE FIX: Use .centroid for the location
    for _, row in gdf.iterrows():
        # Centroid works for Points, Lines, and MultiPolygons
        point = row.geometry.centroid 
        
        folium.CircleMarker(
            location=[point.y, point.x], # Always returns a single lat/lon
            radius=4,
            popup=f"<b>{row['name']}</b><br>{row['fclass']}",
            tooltip=row['name'],
            color="red" if row['fclass'] in ['food_court', 'restaurant'] else "blue",
            fill=True
        ).add_to(m)

    m.save("singapore_fixed_final.html")
    print("SUCCESS: Open singapore_fixed_final.html")

if __name__ == "__main__":
    map_anything()
```
